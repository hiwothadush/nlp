```python
#removeing stopwords
import spacy
import pandas as pd

from spacy.lang.en.stop_words import STOP_WORDS
nlp = spacy.load('en_core_web_sm')
len(STOP_WORDS)
```

```python
text= "I have just started my journy and"
def preprocess(text):
    doc=nlp(text)
    tokens=[token.text for token in doc if not token.is_stop and not token.is_punct ]
    processed_text = " ".join(tokens)
    return processed_text

preprocess(text)
```

```python
#removing all stop owrds andpunctuation
text = '''The Ethiopia national football team,known as the Walias or the Walyas,represents the nation in international football competitions. Governed by the Ethiopian Football Federation (EFF), the team competes in tournaments organized by the Confederation of African Football (CAF) and FIFA. Football in Ethiopia has a rich history dating back to the early 20th century, influenced by the presence of foreign nationals and missionaries. The sport has garnered immense popularity across the country, with passionate fans filling stadiums to support their national team. Ethiopian football continues to evolve, driven by grassroots initiatives and a commitment to nurture young talent, aiming to achieve greater success on the international stage.
'''
doc=nlp(text)
tokens=[token.text for token in doc if not token.is_stop and not token.is_punct ]
mydict = {}
for token in tokens:
  count = tokens.count(token)
  mydict[count] = token

print(mydict)
most_frequnt= max(mydict.keys())
most_frequnt_word = mydict[most_frequnt]
print(most_frequnt_word)
```
